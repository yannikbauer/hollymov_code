```python
color_dict = plot.get_color_dict(color_names=['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'yellow', 'lightblue'])
```

```python
keys_opto = (HmovUnit() & keys_crit & {'hmu_omi_stim < -.2'}).fetch(dj.key)
len(keys_opto)
```

```python
#for key in keys:
key = keys_opto[0]
opto_filts = (SplineLNP()&key).fetch('spl_o')
```

```python
data_fs = 60
filt_len = 20
T = 1 / data_fs
ts = np.arange(-T * filt_len, 0, T)

for key in keys_opto:
    omi = (HmovUnit()&key).fetch1('hmu_omi_e')
    opto_filts = (SplineLNP()&key).fetch('spl_o')
    
    fig, axs = plt.subplots(1,2, figsize=(5,1.5))
    for opto_filt in opto_filts:
        if not (opto_filt == 'None'):
            axs[0].plot(ts, np.squeeze(opto_filt), color=color_dict['blue'], alpha=.5, zorder=1)
        axs[0].hlines(xmin=ts[1], xmax=0, y=0, color='gray', linestyle='--', zorder=0)
        axs[0].set_ylim(-.25, 0.1) 
        axs[0].set_xlabel('Time (s)')
        
    # Plot opto PSTH
    axs2 = (HmovUnit()&key).plot_opto_cond_psth(offsets=[-0.3, 0], optocond=['off-on',  'off-off'], plot_err=False, ax=axs[1], title_info=None, legend=False)
    
    fig.suptitle('{:s} s{:02d} e{:02d} u{:02d}\nOMI={:.2f}'.format(key['m'], key['s'], key['e'], key['u'], omi), y=1.3)
    fig.subplots_adjust(wspace=.5)
    sns.despine()
    
    fig.patch.set_facecolor('white')
    fig.savefig('./plots/{:s}_{:02d}_{:02d}_{:02d}_opto_filt_psth.png'.format(key['m'], key['s'], key['e'], key['u']), dpi=150, bbox_inches='tight', facecolor=fig.get_facecolor(), transparent=False)
```
